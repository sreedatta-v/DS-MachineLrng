```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data_cleaned.csv")
```

```python
data
```

```python
data.shape
```

```python
#Seperating the Independent and dependent Variables
x = data.drop(['Survived'],axis=1)
y = data['Survived']
"""
-> drop() function is used to drop the particluar column/row mentioned in sqr brackets.
-> axis = 1, is used to fro performing oeprations on column basis, axis = 0, for row operations
-> x: Features / Selects all columns excluding particlar column
-> y: Target Variable / Selects only a particlar column."""
```

```python
from sklearn.model_selection import train_test_split
"""
-> Importing train_test_split module from sklearn which basically used to perform spliting the data using this module.
-> test_size: 0.0 ⁓ 1.0, which defines proportion of dataset to be included.
-> train_set/test_set: 0.0 ⁓ 1.0, defines the portion to be added in train/test dataset
-> random_state: It is used to ensure the same randomness inbefore and after the division.
-> shuffle: It is used to indicate weather to shuffle a dataset or not
->Stratify: It is used to ensures that same proportions of different classes present in train and test datasets.
"""
```

```python
train_x,test_x,train_y,test_y = train_test_split(x,y,random_state=101,stratify=y)
```

```python
train_y.value_counts()/len(train_y)
```

```python
test_y.value_counts()/len(test_y)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(train_x,train_y)
```

```python
clf.score(train_x, train_y)
```

```python
clf.score(test_x, test_y)
```
